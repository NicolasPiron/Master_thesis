```python
import numpy as np
import matplotlib.pyplot as plt
import mne
import mne_connectivity
from mne_connectivity import spectral_connectivity_epochs
import pandas as pd
import seaborn as sns

%matplotlib notebook
```

# Fronto-partietal PLV matrix

```python
data = mne.read_epochs('/Users/nicolaspiron/Documents/PULSATION/Python_MNE/preproc/Alpheye_out/data/epochs/S01_r1_epochs.fif')

frequences = {'delta':[0.1, 4],
             'alpha':[8, 12],
             'beta':[13, 30]}

for index_freq, freq in enumerate(range(len(frequences.items()))):
    
    conn_by_idx = []
    fs = 500  # sampling rate (Hz)
    indices = ([4, 6, 39, 41], [20, 22, 57, 59])

    # Frequence chosen for each iteration -> we go in the dict and take the pair of numbers related to the iteration
                             
    fmin = list(frequences.values())[index_freq][0]
    fmax = list(frequences.values())[index_freq][1]

    # It takes the 1st electrode of the 1st list, make a list with only this electrde -> compute the connectivity (PLV) 
    # for this electrode with all the ones if the 2nd list, then do it again with the 2nd electrode and so on. 

    for indice in range(len(indices[0])):
        i_indices = []
        for i in range(len(indices[0])):
            i_indices.append(indices[0][indice])
            total_indices = i_indices, indices[1]

        tuple(total_indices)
        conn_by_idx.append(
            spectral_connectivity_epochs(
                data, method='plv', sfreq=fs, indices=total_indices,
                fmin=fmin, fmax=fmax, faverage=True).get_data()[:, 0])


    # same but for the second list of electrodes

    for indice in range(len(indices[1])):
        i_indices = []
        for i in range(len(indices[1])):
            i_indices.append(indices[1][indice])
            total_indices = indices[0], i_indices

        tuple(total_indices)
        conn_by_idx.append(
            spectral_connectivity_epochs(
                data, method='plv', sfreq=fs, indices=total_indices,
                fmin=fmin, fmax=fmax, faverage=True).get_data()[:, 0])


    # Frontal to frontal

    for indice in range(len(indices[0])):
        i_indices = []
        for i in range(len(indices[0])):
            i_indices.append(indices[0][indice])
            total_indices = i_indices, indices[0]

        tuple(total_indices)
        conn_by_idx.append(
            spectral_connectivity_epochs(
                data, method='plv', sfreq=fs, indices=total_indices,
                fmin=fmin, fmax=fmax, faverage=True).get_data()[:, 0])


    # Parietal to parietal

    for indice in range(len(indices[1])):
        i_indices = []
        for i in range(len(indices[1])):
            i_indices.append(indices[1][indice])
            total_indices = indices[1], i_indices

        tuple(total_indices)
        conn_by_idx.append(
            spectral_connectivity_epochs(
                data, method='plv', sfreq=fs, indices=total_indices,
                fmin=fmin, fmax=fmax, faverage=True).get_data()[:, 0])



    conn_by_idx = np.array(conn_by_idx)

    # Put the lines in the right order
    
    F_to_P = conn_by_idx[:4,:]
    P_to_F = conn_by_idx[4:8,:]
    F_to_F = conn_by_idx[8:12,:]
    P_to_P = conn_by_idx[12:,:]

    F = np.concatenate((F_to_F, F_to_P), axis=1)
    P = np.concatenate((P_to_F, P_to_P), axis=1)
    total_matrix = np.concatenate((F, P), axis=0)

    # Into df to label the rows and columns
    
    df = pd.DataFrame(total_matrix)

    df.columns = ['F3', 'F7', 'F4', 'F8', 'P3', 'P7', 'P4', 'P8']
    df.index = ['F3', 'F7', 'F4', 'F8', 'P3', 'P7', 'P4', 'P8']

    vmin = 0
    vmax = 1

    plt.figure()
    plt.imshow(df, cmap='viridis', vmin=vmin, vmax=vmax)
    plt.xlabel(f'{str(list(frequences.keys())[index_freq])}, {str(list(frequences.values())[index_freq])}')
    plt.xticks(range(len(df.columns)), df.columns)
    plt.yticks(range(len(df.index)), df.index)
    plt.colorbar().set_label('PLV')
    plt.show()
    plt.savefig( '/Users/nicolaspiron/Desktop/' + str(list(frequences.keys())[index_freq]) + '.png')
```

```python
frequences = {'delta':[0.1, 4],
             'alpha':[8, 12],
             'beta':[13, 30]}

for index_freq, freq in enumerate(range(len(frequences.items()))):
    
    conn_by_idx = []
    fs = 500  # sampling rate (Hz)
    indices = ([4, 6, 39, 41], [20, 22, 57, 59])

    # Frequence choson for each iteration -> we go in the dict and take the pair of numbers related to the iteration
                             
    fmin = list(frequences.values())[index_freq][0]
    fmax = list(frequences.values())[index_freq][1]

    # It takes the 1st electrode of the 1st list, make a list with only this electrde -> compute the connectivity (PLV) 
    # for this electrode with all the ones if the 2nd list, then do it again with the 2nd electrode and so on. 

    for indice in range(len(indices[0])):
        i_indices = []
        for i in range(len(indices[0])):
            i_indices.append(indices[0][indice])
            total_indices = i_indices, indices[1]

        tuple(total_indices)
        conn_by_idx.append(
            spectral_connectivity_epochs(
                data, method='plv', sfreq=fs, indices=total_indices,
                fmin=fmin, fmax=fmax, faverage=True).get_data()[:, 0])


    # same but for the second list of electrodes

    for indice in range(len(indices[1])):
        i_indices = []
        for i in range(len(indices[1])):
            i_indices.append(indices[1][indice])
            total_indices = indices[0], i_indices

        tuple(total_indices)
        conn_by_idx.append(
            spectral_connectivity_epochs(
                data, method='plv', sfreq=fs, indices=total_indices,
                fmin=fmin, fmax=fmax, faverage=True).get_data()[:, 0])


    # Frontal to frontal

    for indice in range(len(indices[0])):
        i_indices = []
        for i in range(len(indices[0])):
            i_indices.append(indices[0][indice])
            total_indices = i_indices, indices[0]

        tuple(total_indices)
        conn_by_idx.append(
            spectral_connectivity_epochs(
                data, method='plv', sfreq=fs, indices=total_indices,
                fmin=fmin, fmax=fmax, faverage=True).get_data()[:, 0])


    # Parietal to parietal

    for indice in range(len(indices[1])):
        i_indices = []
        for i in range(len(indices[1])):
            i_indices.append(indices[1][indice])
            total_indices = indices[1], i_indices

        tuple(total_indices)
        conn_by_idx.append(
            spectral_connectivity_epochs(
                data, method='plv', sfreq=fs, indices=total_indices,
                fmin=fmin, fmax=fmax, faverage=True).get_data()[:, 0])



    conn_by_idx = np.array(conn_by_idx)

    F_to_P = conn_by_idx[:4,:]
    P_to_F = conn_by_idx[4:8,:]
    F_to_F = conn_by_idx[8:12,:]
    P_to_P = conn_by_idx[12:,:]

    F = np.concatenate((F_to_F, F_to_P), axis=1)
    P = np.concatenate((P_to_F, P_to_P), axis=1)
    total_matrix = np.concatenate((F, P), axis=0)

    # Reshape the matrix into a 4 by 2 by 4 by 2 array
    submatrices = total_matrix.reshape(4, 2, 4, 2)

    # Average each 2 by 2 submatrix along both dimensions
    averaged_submatrices = submatrices.mean(axis=(1, 3))

    df = pd.DataFrame(averaged_submatrices)

    df.columns = ['Frontal left', 'Frontal right', 'Parietal left', 'Parietal right']
    df.index = ['Frontal left', 'Frontal right', 'Parietal left', 'Parietal right']

    vmin = 0
    vmax = 1

    plt.figure()
    plt.imshow(df, cmap='viridis', vmin=vmin, vmax=vmax)
    plt.xlabel(f'{str(list(frequences.keys())[index_freq])}, 8-12Hz')
    plt.xticks(range(len(df.columns)), df.columns)
    plt.yticks(range(len(df.index)), df.index)
    plt.colorbar().set_label('PLV')
    plt.show()
    plt.savefig( '/Users/nicolaspiron/Desktop/Simple_' + str(list(frequences.keys())[index_freq]) + '.png')
```
